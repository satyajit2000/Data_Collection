```python
import requests
import pandas as pd
import numpy as np
```

```python
request= requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1').json()
```

```python
request
```

```python
popularity=[]
vote_count=[]
original_language=[]
title=[]
vote_average=[]
overview=[]
release_date=[]

for i in request['results']:
    
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    original_language.append(i['original_language'])
    title.append(i['title'])
    vote_average.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])
    
d = {'title':title,'overview':overview, 'release_date':release_date, 'original_language':original_language, 'popularity':popularity, 'vote_count':vote_count, 'vote_average':vote_average} 
df = pd.DataFrame(d)
df
```

```python
finaldf = pd.DataFrame()

for j in range(1,21):
    popularity=[]
    vote_count=[]
    original_language=[]
    title=[]
    vote_average=[]
    overview=[]
    release_date=[]

    for i in request['results']:

        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        title.append(i['title'])
        vote_average.append(i['vote_average'])
        overview.append(i['overview'])
        release_date.append(i['release_date'])

    d = {'title':title,'overview':overview, 'release_date':release_date, 'original_language':original_language, 'popularity':popularity, 'vote_count':vote_count, 'vote_average':vote_average} 
    df = pd.DataFrame(d)
    
    finaldf = finaldf.append(df, ignore_index=True)
    
```
